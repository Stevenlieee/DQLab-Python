#List (array)

```python
#           0         1         2       3   4      5    DARI DEPAN
#          -6         -5        -4    -3   -2      -1    DARI BELAKANG
fruits = ["Apple", "Banana", "Cherry", 4, 1.5] # orange
print(fruits[0]) #Output: Apple
buah1 = fruits[0]
print(buah1)
fruits.append("Orange") #Add a new item
print(fruits)
print(fruits[5])
print(fruits[-1])
print(fruits[0::1])
print(fruits[-1::-1])
print(fruits[0:4:1])
print(fruits[-1::-2])
```
